# echeveria quickstart

This notebook runs on the base install with no data: it synthesizes a plant image, analyzes it, and shows the overlay and the pigment saliency map. Every output is an RGB proxy, so treat it as indicative.

```python
import numpy as np

from phytovision import Pipeline
from phytovision.visualize import render_overlay, render_saliency_overlay


def synthetic_plant(size=256, colour=(0.62, 0.5, 0.12)):
    """A yellowing plant on a dark background, so the score reads as stressed."""
    yy, xx = np.mgrid[0:size, 0:size]
    img = np.ones((size, size, 3), np.float32) * 0.1
    img[(yy - size / 2) ** 2 + (xx - size / 2) ** 2 <= (size * 0.32) ** 2] = colour
    return img


image = synthetic_plant()
```

```python
pipeline = Pipeline.default()
report = pipeline.analyze(image)
report.summary()
```

The overlay tints the plant by the stress score; the saliency map shows which colour pixels drove that score. Red raises the score, green lowers it.

```python
render_overlay(image, report)
```

```python
render_saliency_overlay(image, report, pipeline.model)
```
